```python
# We need a function to convert EncodedPixels into mask

def rle2mask(mask_rle, shape=(1600,256)):
    s = mask_rle.split()
    starts, lengths = [np.asarray(x, dtype=int) for x in (s[0:][::2], s[1:][::2])]
    starts -= 1
    ends = starts + lengths
    img = np.zeros(shape[0]*shape[1], dtype=np.uint8)
    for lo, hi in zip(starts, ends):
        img[lo:hi] = 1
    return img.reshape(shape).T

def mask2rle(img):
    pixels= img.T.flatten()
    pixels = np.concatenate([[0], pixels, [0]])
    runs = np.where(pixels[1:] != pixels[:-1])[0] + 1
    runs[1::2] -= runs[::2]
    return ' '.join(str(x) for x in runs)
```

```python
# Metrics
# For image segmentation

def dice_coef(y_true, y_pred, smooth=K.epsilon()):

    y_true_f = K.flatten(y_true)
    y_pred_f = K.flatten(y_pred)
    intersection = K.sum(y_true_f * y_pred_f)
    return (2. * intersection + smooth) / (K.sum(y_true_f) + K.sum(y_pred_f) + smooth)


# For clasification
def recall_m(y_true, y_pred):
    true_positives = K.sum(K.round(K.clip(y_true * y_pred, 0, 1))) # calculates number of true positives
    possible_positives = K.sum(K.round(K.clip(y_true, 0, 1)))      # calculates number of actual positives
    recall = true_positives / (possible_positives + K.epsilon())   # K.epsilon takes care of non-zero divisions
    return recall

def precision_m(y_true, y_pred):

    true_positives = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))  # calculates number of true positives
    predicted_positives = K.sum(K.round(K.clip(y_pred, 0, 1)))      # calculates number of predicted positives   
    precision = true_positives /(predicted_positives + K.epsilon()) # K.epsilon takes care of non-zero divisions
    return precision
    
def f1_score_m(y_true, y_pred):
    precision = precision_m(y_true, y_pred)  # calls precision metric and takes the score of precision of the batch
    recall = recall_m(y_true, y_pred)        # calls recall metric and takes the score of precision of the batch
    return 2*((precision*recall)/(precision+recall+K.epsilon()))

dependencies = {
    'recall_m':recall_m,
    'precision_m':precision_m,
    'dice_coef':dice_coef,
    'f1_score_m':f1_score_m,
    'dice_loss':sm.losses.dice_loss
}
```

# Binary classsification

```python
#data preparation for binary classification
X_train_binary = X_train[['ImageId','hasDefect']]
X_val_binary = X_val[['ImageId','hasDefect']]
X_test_binary = X_test[['ImageId','hasDefect']]

print(X_train.shape, X_val.shape, X_test.shape)
print(X_train_binary.shape, X_val_binary.shape, X_test_binary.shape)
```

```python

# DataGenerator for the binary classification model with image augmentations

train_DataGenerator_1 = ImageDataGenerator(rescale=1./255., shear_range=0.2, zoom_range=0.05, rotation_range=5,
                           width_shift_range=0.2, height_shift_range=0.2, horizontal_flip=True, vertical_flip=True)

test_DataGenerator_1 = ImageDataGenerator(rescale=1./255)

train_generator = train_DataGenerator_1.flow_from_dataframe(
        dataframe=X_train_binary.astype(str),
        directory=train_path,
        x_col="ImageId",
        y_col="hasDefect",
        target_size=(256,512),
        batch_size=16,
        class_mode='binary')

validation_generator = test_DataGenerator_1.flow_from_dataframe(
        dataframe=X_val_binary.astype(str),
        directory=train_path,
        x_col="ImageId",
        y_col="hasDefect",
        target_size=(256,512),
        batch_size=16,
        class_mode='binary')
```

```python
# Using a pretrained model from keras for classification: 
# Selecting Xception pretrained model


base_model = keras.applications.xception.Xception(include_top = False, input_shape = (256,512,3))

# add a global spatial average pooling layer
x = base_model.output
x = GlobalAveragePooling2D()(x)

# let's add a fully-connected layer
x = Dense(1024, activation='relu')(x)
x = BatchNormalization()(x)
x = Dropout(0.3)(x)

x = Dense(512, activation='relu')(x)
x = BatchNormalization()(x)
x = Dropout(0.3)(x)

x = Dense(64, activation='relu')(x)

# and the prediction layer
predictions = Dense(1, activation='sigmoid')(x)

# this is the model we will train
model = Model(inputs=base_model.input, outputs=predictions)
model.summary()
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy',metrics=['acc',f1_score_m,precision_m,recall_m])
logdir = "/content/drive/My Drive/ml project result/logs/scalars/" + datetime.now().strftime("%Y%m%d-%H%M%S")+'_binary'
file_writer = tf.summary.create_file_writer(logdir + "/metrics")
mc = ModelCheckpoint('/content/drive/My Drive/ml project result/model/binary.h5', monitor='val_f1_score_m', mode='max', verbose=1, save_best_only=True)
history = model.fit(train_generator, validation_data = validation_generator, epochs = 15, verbose=1, callbacks = [mc])
file_writer.close()
```

```python
#custome class for data generation with augmentation
class train_DataGenerator_3(tf.keras.utils.Sequence): 
    def __init__(self, df, batch_size=8,  shuffle=True, 
                 preprocess=None, info={}):
        super().__init__()
        self.df = df
        self.shuffle = shuffle
        self.batch_size = batch_size
        self.preprocess = preprocess
        self.info = info
        self.data_path = train_path
        self.on_epoch_end()

    def __len__(self):
        return int(np.floor(len(self.df) / self.batch_size))
    
    def on_epoch_end(self):
        self.indexes = np.arange(len(self.df))
        if self.shuffle == True:
            np.random.shuffle(self.indexes)
    
    def __getitem__(self, index): 
        X = np.empty((self.batch_size,256,1600,3),dtype=np.float32)
        X1 = np.empty((self.batch_size,256,1600,3),dtype=np.float32)

        y = np.empty((self.batch_size,256,1600,1),dtype=np.float32)
        y1 = np.empty((self.batch_size,256,1600,1),dtype=np.float32)

        indexes = self.indexes[index*self.batch_size:(index+1)*self.batch_size]
        for i,f in enumerate(self.df['ImageId'].iloc[indexes]):
            self.info[index*self.batch_size+i]=f
            X[i,] = Image.open(self.data_path + '/' + f)          
            y[i,:,:,0] = rle2mask(self.df['EncodedPixels'].iloc[indexes[i]])

        # generate some random image augmentations
        augment = np.random.random()
        if augment>0.35:
            in_gen1 = ImageDataGenerator()
            augment1 = np.random.random()
            augment2 = np.random.random()
            augment3 = np.random.random()
            augment4 = np.random.random()
            augment5 = np.random.random()
            augment6 = np.random.random()

            args = dict(tx = 0, ty = 0, zx = 1.0, zy= 1.0, flip_horizontal = False, flip_vertical = False)

            if augment1>0.5:
                args.update({'tx':50})

            if augment2>0.5:
                args.update({'ty':25})

            if augment3>0.5:
                args.update({'zx':0.9})

            if augment4>0.5:
                args.update({'zy':0.9})

            if augment5>0.5:
                args.update({'flip_horizontal' : True})

            if augment6>0.5:
                args.update({'flip_vertical' : True})

            for i,h in enumerate(X):
                X1[i] = in_gen1.apply_transform(h, transform_parameters = args)
            for i,g in enumerate(y):
                y1[i] = in_gen1.apply_transform(g, transform_parameters = args)
            return X1, y1
        else:
            return X, y
```

```python
# method to identify class labels
def get_classid(y):
  y = y.copy()
  temp = []
  for i in range(len(y)):
    if y.loc[i,'hasDefect'] >= 0.8:
      if y.loc[i,'hasDefect_1'] >= 0.5:
        temp.append((y.loc[i,'ImageId'], 1))
      if y.loc[i,'hasDefect_2'] >= 0.5:
        temp.append((y.loc[i,'ImageId'], 2))
      if y.loc[i,'hasDefect_3'] >= 0.5:
        temp.append((y.loc[i,'ImageId'], 3))
      if y.loc[i,'hasDefect_4'] >= 0.5:
        temp.append((y.loc[i,'ImageId'], 4))
      if y.loc[i,'hasDefect_1'] < 0.5 and y.loc[i,'hasDefect_2'] < 0.5 and y.loc[i,'hasDefect_3'] < 0.5 and y.loc[i,'hasDefect_4'] < 0.5:
        temp.append((y.loc[i,'ImageId'], 0))
    else:
      temp.append((y.loc[i,'ImageId'], 0))
  return pd.DataFrame(temp, columns = ['ImageId','ClassId'])
```

```python
# method to predict class baesd on classification
def predict_classification(X, path):
  X = X.reset_index().drop('index',axis=1)
  data_generator = ImageDataGenerator(rescale=1./255).flow_from_dataframe(dataframe=X, directory=path, 
                                                                            x_col="ImageId", class_mode = None, 
                                                                            target_size=(256,800), batch_size=1, shuffle=False)
  
  data_preds_binary = binary_model.predict(data_generator,verbose=0)
  data_preds_multi_label = multi_model.predict(data_generator,verbose=0)
  data_classification = pd.DataFrame(data_preds_multi_label, columns = ['hasDefect_1','hasDefect_2','hasDefect_3','hasDefect_4'])
  data_classification['hasDefect'] = data_preds_binary
  data_classification['ImageId'] = X['ImageId']
  image_classId = get_classid(data_classification[['ImageId','hasDefect','hasDefect_1','hasDefect_2','hasDefect_3','hasDefect_4']])
  return image_classId
```

```python
# method to plot confusion matrix
def plot_confucion(y):
  clf_report = pd.DataFrame(classification_report(y.iloc[:,1],y.iloc[:,2], output_dict=True))
  print(f"Accuracy Score: {accuracy_score(y.iloc[:,1],y.iloc[:,2]) * 100:.2f}%")
  print("___________________________________________________________________________")
  print(f"CLASSIFICATION REPORT:\n{clf_report}")
  print("___________________________________________________________________________")
  cm = confusion_matrix(y.iloc[:,1],y.iloc[:,2])
  plt.figure(figsize=(10,8))
  sns.heatmap(cm, annot=True, cmap='Blues',fmt='g')
  plt.title('CONFUSION MATRIX')
  plt.ylabel('Actal Values')
  plt.xlabel('Predicted Values')
  plt.show()
```

```python
# method to predict segmentation with batches
def pred_segmentation(X):
    X = X.reset_index().drop('index',axis=1)
    tmp=[]
    loop_num = 50
    for j in range((len(X)//loop_num)+1):
    
      test_dataf = X[loop_num*j:loop_num*j+loop_num]

      test_batches =  test_DataGenerator(test_dataf)
      test_preds_1 = model_segment1.predict(test_batches,verbose=0)
      test_preds_2 = model_segment2.predict(test_batches,verbose=0)
      test_preds_3 = model_segment3.predict(test_batches,verbose=0)
      test_preds_4 = model_segment4.predict(test_batches,verbose=0)
      for i in range(len(test_preds_1)):
        ep1 = mask2rle(np.array((Image.fromarray((test_preds_1[i][:,:,0])>=0.5))).astype(int))
        ep2 = mask2rle(np.array((Image.fromarray((test_preds_2[i][:,:,0])>=0.5))).astype(int))
        ep3 = mask2rle(np.array((Image.fromarray((test_preds_3[i][:,:,0])>=0.5))).astype(int))
        ep4 = mask2rle(np.array((Image.fromarray((test_preds_4[i][:,:,0])>=0.5))).astype(int))
        tmp.append([test_dataf.ImageId.iloc[i],ep1,ep2,ep3,ep4])

    return pd.DataFrame(tmp,columns=['ImageId','EncodedPixels_1','EncodedPixels_2','EncodedPixels_3','EncodedPixels_4'])
```
